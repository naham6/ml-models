```python
X = data.drop(columns=['Price', 'Address'])
Y = data.Price
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.score(xtest,ytest)
```
